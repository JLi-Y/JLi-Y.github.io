- Intro

The data that is being concluded would cover the KKK activities that happened in Colorado in the 1920s. It has included numerical data for the population of active members by the three ledgers. During that time the KKK was surprisingly influential in Colorado, even taking a role in state politics. In the ledgers, we’re able to find details like names, addresses, and membership fees. Making us able to paint a picture of how the organization ran back then. Nowadays, History Colorado has digitalized these documents, making these data public for educational and informative usage. 

- Who collected, processed, and made this information available?

During the 1920s, local KKK groups; called Klaverns, made these ledgers to record member names, dues, and other organizational details. Over time, these books were passed from person to person or sold, until they finally reached History Colorado. Recognizing their historical importance, archivists from History Colorado took care of the ledgers and eventually digitized them making them available to the public. By scanning the pages and creating a website, History Colorado allowed people to research both their family histories and the KKK’s impact on Colorado politics.

- Where and when was this information collected, processed, and made available?

At first, when the KKK held most of the power in Colorado, these ledgers were put together in different parts of the state. Due to the popularity of the KKK back then, they have recorded most of their members in these books. After the KKK started to lose influence in the late 1920s, some ledgers went missing or stayed in people’s private collections for many years. Eventually, History Colorado gathered as many as it could find and preserved them in the later part of the 20th century. The major push to digitize them happened more recently, in the 2010s, so that anyone could easily access these records online and learn about the history of the KKK in Colorado.

- How was this information collected, processed, and made available?

Local KKK leaders made these records to keep track of new members, list dues payments, and note each person’s status; if they were in good standing. When History Colorado got these ledgers, archivists had to be very careful with them since some pages were torn or in poor shape. They cleaned and fixed the pages, placed them in protective covers, and stored them following professional archival rules. After that, they digitized the documents by scanning or photographing each page in high resolution. Staff and volunteers also created a partial index with basic details such as page numbers and dates to help people search the digital images. Lastly, History Colorado posted everything on their website so researchers, students, and the public could access them easily.

- Why was this information collected, processed, and made available?

From the Klan’s point of view, these ledgers were a straightforward way to handle money and track members in the 1920s. Keeping detailed accounts helped them stay organized, finance their activities, and gain political power. On the other hand, History Colorado’s main aim is different. By preserving and digitizing these documents, they offer an important firsthand look at the KKK’s presence and impact at the local level. Historians can see how many people paid dues and took part, while people interested in family history might find surprising connections. More importantly, sharing these materials publicly helps us confront the history of racism and white supremacy. Even though these records can be uncomfortable, making them available gives a clearer view of how these harmful ideas shaped and still affect society.

- How is this data stored today? How did you access it?

Currently, the original ledger books are stored in secure facilities at History Colorado, where they’re kept in climate-controlled conditions to stop them from falling apart. Digitally, high-resolution scans of these pages are saved on servers and backed up both on-site and possibly online, so they don’t get lost if something goes wrong. People can see the scanned images on History Colorado’s website, which usually includes basic details like page numbers. Even though there’s no full transcription of every page yet, the site does have some search tools that help you find certain names or keywords. If anyone wants to look at the actual physical ledgers, they can set up a visit with the archives, but for quick or causal research, the online version is the easiest option.

- What kind of data is it? What information does it contain?

The KKK ledgers include both qualitative and quantitative data. On the quantitative side, you’ll find numbers and figures like how many members there were, how much they paid in dues, and the dates they paid. This kind of information can be counted and compared. On the qualitative side, there are written details like names, addresses, or short notes about members’ statuses. These bits of text can’t be turned into simple numbers as easily, but they still provide important context about who these people were and how the Klan operated.

- What are the limitations or weaknesses of the dataset? What is missing?

Although these ledgers are critically important for researching the KKK, they are not exhaustive. Certain volumes appear to have been lost, leaving gaps in the historical record for specific towns or periods. Moreover, the handwritten entries can be difficult to interpret, which may result in transcription errors or incomplete digital indexes. The ledgers also focus primarily on individuals who formally paid dues, meaning that informal supporters or collaborators might be absent. In addition, appearing in a ledger does not necessarily clarify how an individual participated in the organization’s racist or antisemitic activities. Researchers must further account for ethical considerations, as identifying KKK members can raise privacy concerns or reopen painful family histories.

- What social, cultural, or political contexts should a user keep in mind while using the dataset?

People reading or using these ledgers should remember that the KKK was a white supremacist group built on racism, xenophobia, and antisemitism. In 1920s Colorado, the Klan helped elect politicians and judges who then shaped laws and policies targeting Catholics, Jews, and Black communities. From a Data Feminism viewpoint, it’s important to realize that the KKK collected data in a biased way, mainly focusing on money and memberships to push their hateful goals. Also, these records don’t include the perspectives of those the Klan harmed. Knowing this bigger picture helps us see the ledgers as a starting point for deeper research, rather than the whole story on their own.
